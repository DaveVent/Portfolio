# Exemple of Support Vector Machines (classification)

```python
# Packages needed
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
# Import the iris dataset
iris = datasets.load_iris()

# Take the first two features and the target variable
X = iris.data[:, :2]
y = iris.target


# Fit a SVC model with linear kernel
model = svm.SVC(kernel="linear", C=1).fit(X,y)

# Set-up 1x2 grid for plotting
fig, ax = plt.subplots(1,2, figsize=[15, 7])

# ALL POINTS 

# Get the first and the second feature
X0, X1 = X[:, 0], X[:, 1]

# Plot decision boundaries and data points
disp = DecisionBoundaryDisplay.from_estimator(
    model,
    X,
    response_method="predict",
    cmap="ocean",
    alpha=0.8,
    ax=ax[0],
    xlabel=iris.feature_names[0],
    ylabel=iris.feature_names[1],
)
ax[0].scatter(X0, X1, c=y, cmap="ocean", s=20, edgecolors="k")
ax[0].set_xticks(())
ax[0].set_yticks(())
ax[0].set_title("SVC")


# SUPPORT VECTORS ONLY

# Filter the first and the second feature for the support vectors only
X0, X1 = X[model.support_][:, 0], X[model.support_][:, 1]

# Plot decision boundaries and data points
disp = DecisionBoundaryDisplay.from_estimator(
    model,
    X,
    response_method="predict",
    cmap=plt.cm.ocean,
    alpha=0.8,
    ax=ax[1],
    xlabel=iris.feature_names[0],
    ylabel=iris.feature_names[1],
)
ax[1].scatter(X0, X1, c=y[model.support_], cmap=plt.cm.ocean, s=20, edgecolors="k")
ax[1].set_xticks(())
ax[1].set_yticks(())
ax[1].set_title("SVC (support vectors only)")


# Show the plot
plt.show()
```
